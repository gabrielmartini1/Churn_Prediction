##**1. Introdução**

<p align="justify">O setor de telecomunicações é um dos mais promissores e crescentes do mundo. Nos últimos anos, tem passado por um desenvolvimento exponencial, afetando cerca de 90% da população mundial. Esse crescimento se deve principalmente aos avanços tecnológicos e novos serviços de telecomunicações, que geram cada vez mais dados e se tornam ativos valiosos para empresas e organizações.</p>

<p align="justify">No entanto, a competição no mercado de telecomunicações é acirrada. Com múltiplas empresas oferecendo serviços similares a preços restritos, a rotatividade de clientes é uma preocupação central para as receitas das empresas.</p>

<p align="justify">Fatores como o aumento dos valores dos planos, a falta ou a baixa qualidade de suporte ao cliente, experiências negativas do cliente, concorrentes com melhores condições contratuais, entre outros,  podem levar os clientes à rotatividade.</p>

<p align="justify">Para mitigar esse problema, é fundamental prever quais clientes estão em risco de rotatividade, sendo essa uma métrica importante para a saúde da empresa. Vários pesquisadores estão trabalhando em técnicas para prever a rotatividade dos clientes de telecomunicações, já que a indústria tem perdido mais de 25% de seus clientes a cada ano, o que resulta em uma grande perda de receita. Além disso, adquirir um novo cliente custa de 5 a 10 vezes mais do que manter um cliente atual.</p>

<p align="justify">Portanto, é crucial manter os clientes existentes e evitar a rotatividade.
Churn rate ou taxa de abandono é o indicador que mede o número de clientes que cancelam o produto ou serviço. Quando a taxa de Churn é muito alta, isso significa que há problemas nas estratégias da empresa.</p>

<p align="justify">Para prever a taxa de churn do cliente e tomar as medidas preventivas, é necessário coletar e analisar as informações sobre o comportamento do cliente, como o intervalo de compras, cancelamento  de serviços, histórico de pesquisa de satisfação (NPS).</p>

<p align="justify">Saber com antecedência quais clientes podem desistir em breve dos serviços prestados e desenvolver estratégias eficientes de retenção, podem melhorar o LTV (Lifetime Value) da empresa.</p>

<p align="justify">LTV ou Lifetime Value  é uma métrica que calcula o lucro líquido ao longo do ciclo de vida do cliente junto à empresa. Esse indicador, é muito importante para os negócios de receita recorrente, sendo fundamental para a empresa saber o valor médio gasto por sua base de clientes e por quanto tempo eles costumam manter o relacionamento com a marca.</p>

<p align="justify">O LTV é uma métrica focada na retenção de clientes, ajudando a reduzir os custos de aquisição. O CAC (Custo de Aquisição de Cliente) representa o investimento que uma empresa precisa fazer para adquirir um novo cliente. Isso inclui todos os custos envolvidos no processo de aquisição, como publicidade, marketing, vendas, suporte e outros.</p>

<p align="justify">O CAC é importante porque indica o quanto uma empresa está investindo para adquirir novos clientes e, consequentemente, quanto precisa gerar de receita para se pagar e obter lucro.</p>

<p align="justify">O CAC e o LTV estão diretamente relacionados, já que para uma empresa ser lucrativa, o valor do LTV deve ser maior do que o CAC. Ou seja, o custo de aquisição de um cliente deve ser menor do que o valor que ele pode gerar de receita ao longo do tempo para que a empresa obtenha lucro com esse cliente.</p>

<p align="justify">Nesse sentido, o CAC e o LTV são métricas importantes para avaliar a efetividade dessas estratégias e identificar oportunidades de melhoria. Empresas que conseguem reduzir o CAC e aumentar o LTV tendem a ter um modelo de negócio mais sustentável e rentável a longo prazo.</p>

## **4. Pré-processamento de Dados**

<p align="justify">O pré-processamento de dados é uma etapa fundamental em todos os projetos de ciências de dados e engenharia de machine learning, onde os dados brutos são transformados em um formato que possa ser facilmente analisado e modelado por algorítmos de aprendizado de máquina.</p>

<p align="justify">Esta etapa de pré-processamento de dados envolve a execução de uma série de tarefas, tais como: a limpeza dos dados, redução, integração de dados, criação ou transformação de novas variáveis, seleção de dados, balanceamento de classes, divisão do conjunto de dados, padronização e normalização dos dados.</p>

<p align="justify">A limpeza envolve a identificação e remoção de dados duplicados, inconsistências, erros e outliers, enquanto a transformação pode incluir normalização, discretização e codificação de variáveis categóricas.</p>

<p align="justify">Redução de dimensionalidade é uma técnica para reduzir a complexidade do modelo, como por exemplo, através da seleção de características relevantes e descarte de variáveis redundantes para melhorar a performance do modelo.</p>

<p align="justify">A integração envolve a combinação de dados de múltiplas fontes em um único conjunto de dados. Já a seleção de dados é a escolha das variáveis mais relevantes para o modelo preditivo.</p>

<p align="justify">A criação ou a transformação de novas variáveis (features engineering), ocorre a partir das variáveis existentes, para que possam ter maior relevância para o modelo.</p>

<p align="justify">O balanceamento ocorre quando a classe alvo do modelo está desbalanceada. É necessário aplicar técnicas para balanceá-la, a fim de evitar um modelo enviesado.</p>

<p align="justify">A divisão do conjunto de dados, refere-se a separação dos dados em treino, validação e teste, a fim de avaliar o modelo de forma adequada.</p>

<p align="justify">A padronização e a normalização são técnicas de "feature scaling". Essas técnicas são usadas para colocar as variáveis (ou features) em uma mesma escala numérica. O objetivo da "feature scaling" é garantir que todas as variáveis tenham um peso igual no modelo de machine learning.</p>

<p align="justify">O objetivo do pré-processamento de dados é melhorar a qualidade dos dados, prepará-los para serem usados em modelos de aprendizado de máquina e garantir que os resultados da análise sejam precisos e confiáveis. <p align="justify">Somente a variável 'Churn Reason', apresentou 5.174 (73,4%) registros ausentes no conjunto de dados. Para lidarmos com esse problema de ausência de dados no dataset, vamos utilizar algorítmos específicos de imputação para tratar as colunas numéricas, categóricas e categóricas numéricas, conforme código abaixo:</p>

```python
# Criar uma cópia do dataset
df_clean = df.copy()

# Selecionar as colunas do tipo 'object' do dataset
cols_categoricas = df_clean.select_dtypes(include=['object']).columns

# Selecionar as colunas do tipo 'float' e 'int' do dataset
cols_numericas = df_clean.select_dtypes(include=[float, int]).columns

# Imputar os valores faltantes nas colunas numéricas com KNNImputer
imputer_numericas = KNNImputer(n_neighbors=5)
df_clean[cols_numericas] = imputer_numericas.fit_transform(df_clean[cols_numericas])

# Imputar os valores faltantes nas colunas categóricas e categóricas numéricas com SimpleImputer
imputer_categoricas = SimpleImputer(strategy='most_frequent')
cols_categoricas_numericas = ['Churn Reason', 'Senior Citizen', 'Dependents', 'Phone Service', 'Multiple Lines', 'Online Security',
                              'Online Backup', 'Device Protection', 'Tech Support', 'Streaming TV', 'Streaming Movies']
df_clean[cols_categoricas_numericas] = imputer_categoricas.fit_transform(df_clean[cols_categoricas_numericas])
df_clean[cols_categoricas] = df_clean[cols_categoricas].fillna(value='unknown')
```

<p align="justify">O código acima fez o tratamento de valores ausentes (NaN) no dataset utilizando o pacote "sklearn" do Python com as classes de imputação "KNNImputer" e "SimpleImputer".</p>

<p align="justify">A classe de imputação "KNNImputer" utiliza a técnica de K-vizinhos mais próximos (5 vizinhos) para inferir os valores ausentes com base nos valores das observações mais próximas. Esse tratamento ocorre sobre as colunas numéricas.</p>

<p align="justify">A classe "SimpleImputer" utiliza a técnica de imputação simples sobre as colunas categóricas e categóricas numéricas. O parâmetro "strategy" é definido como "most_frequent", o que significa que a imputação é baseada no valor mais frequente na coluna.</p> 

<p align="justify">O método "fillna( )" substitui quaisquer valores ausentes (NaN) restantes nas colunas categóricas por ("unknown"). Isso garante que não haja valores ausentes restantes no conjunto de dados após o tratamento.</p>

###**4.6 Dividindo o dataset entre treino e validação**

<p align="justify"><p align="justify">A divisão do conjunto de dados em treinamento e validação é feita para que possamos avaliar a capacidade do modelo de aprender padrões úteis a partir dos dados de treinamento e aplicar esses padrões a novos dados.</p> 

<p align="justify">O conjunto de dados de treinamento é usado para ajustar os parâmetros do modelo de aprendizado de máquina, ou seja, para fazer com que o modelo "aprenda" a relação entre as variáveis de entrada e de saída, enquanto que o conjunto de validação é usado para avaliar a qualidade do modelo.

###**4.7 Realizando o balanceamento dos dados**

<p align="justify">Quando a proporção de uma classe é significamente maior ou menor que a outra, podemos utilizar técnicas de reamostragem para equilibrar as proporções.</p>

<p align="justify">Se a classe minoritária for uma pequena proporção do conjunto de dados, podemos usar a técnica de "under-sampling" para equilibrar as proporções. O "under-sampling" envolve a remoção aleatória de exemplos da classe majoritária para igualar a proporção da classe minoritária. No entanto, o "under-sampling" pode levar à perda de informações importantes, especialmente se o conjunto de dados original já for pequeno.</p>

<p align="justify">Por outro lado, se o conjunto de dados possuir uma grande diferença entre as proporções das classes, o uso da técnica de "over-sampling" com SMOTE (Synthetic Minority Over-sampling Technique) pode ser uma boa opção. O SMOTE é uma técnica de "over-sampling" que gera novos exemplos sintéticos da classe minoritária, a fim de aumentar a sua representação no conjunto de dados. Isso pode ajudar a melhorar a precisão do modelo, especialmente se houver poucos exemplos da classe minoritária no conjunto de dados.</p>

<p align="justify">Considerando que a classe minoritária (churn) representa cerca de 26,54% do conjunto de dados, vamos aplicar o uso da técnica de SMOTE com "over-sampling", conforme código e output abaixo:</p>

```python
# Aplicando a técnica de SMOTE com over-sampling nos dados de treinamento
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X_train, y_train)

# Ver o balanceamento das classes
print(pd.Series(y_smote).value_counts())
```

### **4.8 Padronização e Normalização das variáveis (Feature Scaling)**

### **4.8.1 Técnica de Padronização**
<p align="justify">A técnica de padronização (Standardization ou Z-score normalization) é uma técnica de pré-processamento de dados que transforma as variáveis para que tenham média igual a zero e desvio padrão igual a um. Isso significa que os valores transformados estarão centrados na média da variável e terão a mesma escala (desvio padrão) da variável original.</p>

<p align="justify">A padronização é mais adequada para variáveis que têm uma distribuição normal ou Gaussiana e não estão limitadas a um intervalo específico. É amplamente utilizada em algorítmos que consideram a escala das variáveis, como a Regressão Linear e a Regressão Logística. A fórmula para a padronização é dada por: X_stand = (X - mean(X)) / std(X), onde X é o valor da variável, mean(X) é a média da variável e std(X) é o desvio padrão da variável.</p>

### **4.8.2 Técnica de Normalização**
<p align="justify">A técnica de normalização (Min-Max Scaling) é uma técnica que consiste em transformar os valores das variáveis para que fiquem no intervalo de 0 a 1. A fórmula para a normalização é dada por: X_norm = (X - X_min) / (X_max - X_min), onde X é o valor da variável, X_min é o valor mínimo da variável e X_max é o valor máximo da variável.</p>

<p align="justify">Essa técnica pode ser útil quando se está trabalhando com algorítmos que consideram a escala das variáveis, como é o caso dos algorítmos de Regressão Linear, Regressão Logística, SVM, KNN, entre outros.</p>

### **4.8.3 Escolha da técnica para o dataset**

<p align="justify">O quadrante III (FN - False Negative), representa o número de instâncias que foram incorretamente classificadas como negativas. <p align="justify">Portanto, para o problema de churn, é importante encontrarmos um equilíbrio entre a precisão e o recall do modelo. Um modelo com alta precisão pode ajudar a evitar gastos desnecessários da empresa, mas pode deixar de identificar alguns clientes que realmente vão cancelar o serviço. Um modelo com alto recall pode identificar a maioria dos clientes que vão cancelar o serviço, mas pode levar a mais gastos da empresa para reter clientes que não vão cancelar.</p>

<p align="justify">Para este projeto, vamos considerar a F1-score como o melhor indicador do relatório de classificação, pois ele representa tanto a precisão quanto o recall do modelo. A precisão mede a proporção de verdadeiros positivos (TP) em relação ao número total de classificações positivas (TP + FP), enquanto o recall mede a proporção de verdadeiros positivos (TP) em relação ao número total de exemplos positivos (TP + FN).</p>

<p align="justify">A F1-score é a média harmônica entre precisão e recall (F1-score = 2 * (precisão * recall) / (precisão + recall)), e seu valor varia de 0 a 1, sendo que um valor mais próximo de 1 indica um modelo com melhor desempenho.</p>

<p align="justify">Outro indicador que podemos utilizar para encontrarmos um ponto de equilíbrio entre a precisão e o recall, é a área sob a curva (AUC), um indicador que avalia o desempenho geral do modelo.</p>

###**6.2.1 Técnica de Validação Cruzada (cross-validation)**

<p align="justify">Para fins de estudo, já testamos o primeiro modelo de machine learning, através da técnica de "over-sampling" com SMOTE sobre o conjunto de dados. Entretanto, mesmo que o primeiro modelo tenha apresentado bom desempenho nos dados de teste, não há garantia de que ele terá um bom desempenho em novos dados.</p>

<p align="justify">Portanto, vamos testar um segundo modelo de machine learning, através da técnica de validação cruzada (cross-validation), que vai avaliar a capacidade do modelo em generalizar o desempenho para novos dados.</p> 

<p align="justify">A validação cruzada divide o conjunto de dados em k partes iguais e executa o modelo k vezes, usando cada parte como conjunto de teste uma vez e as k-1 partes restantes como conjunto de treinamento. Isso permite que todos os dados sejam usados tanto para treinamento quanto para teste, e as métricas de desempenho são calculadas com base em uma média das k execuções.

<p align="justify">O uso da validação cruzada permite uma avaliação mais robusta do desempenho do modelo e ajuda a detectar possíveis problemas de sobreajuste ou subajuste. A seguir, o código do segundo modelo de teste de machine learning:</p>


```python
# Técnica de cross-validation
from sklearn.model_selection import cross_validate

# Definindo o modelo de Regressão Logística
clf = LogisticRegression()

# Definindo o objeto cv com 5 folds
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Gerando as previsões e as classes verdadeiras para cada fold da validação cruzada
y_pred = cross_val_predict(clf, X, y, cv=cv)
y_true = y

# Calculando as métricas de avaliação para cada fold
scoring = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']
scores = cross_validate(clf, X, y, cv=cv, scoring=scoring)

# Imprimindo as métricas de avaliação para cada fold
df_scores = pd.DataFrame(scores)
df_scores = df_scores.drop(columns=['fit_time', 'score_time'])
df_scores = df_scores.apply(lambda x: round(x, 4))
print(df_scores)
print('')

# Imprimindo a média e o desvio padrão de cada métrica
print('Acurácia média:', "{:.2f}".format(scores['test_accuracy'].mean()), '\nDesvio padrão: ', "{:.2f}".format(scores['test_accuracy'].std()))

# Calculando a matriz de confusão consolidada para todos os folds
conf_mat = confusion_matrix(y_true, y_pred, normalize='true')

# Plotando a matriz de confusão consolidada
skplt.metrics.plot_confusion_matrix(y_true, y_pred, normalize='true')
plt.show()
```

<p align="justify">Com a técnica de validação cruzada (cross-validation), obtivemos um desempenho médio da acurácia de 0,89, com desvio padrão de 0,01, o que indica uma boa consistência nos resultados entre os diferentes "folds". O modelo também obteve um bom desempenho nos diferentes indicadores, como a métrica de AUC que teve uma média acima de 0,95 em todos os "folds" </p>

<p align="justify">Os "folds" são as partições que dividem o conjunto de dados original em subconjuntos para a validação cruzada. Por exemplo, em uma validação cruzada com 5 folds, o conjunto de dados é dividido em 5 partes iguais (ou o mais próximo possível disso), chamadas de "folds". Em cada iteração do processo de validação cruzada, um dos "folds" é usado como conjunto de teste e os outros 4 "folds" são usados como conjunto de treinamento. O processo é repetido até que todos os "folds" tenham sido usados como conjunto de teste uma vez. Essa técnica é utilizada para avaliar o desempenho de um modelo de machine learning de forma mais robusta e evitar overfitting.</p>

<p align="justify">Overfitting é um problema comum em machine learning que ocorre quando um modelo se ajusta muito bem aos dados de treinamento e, consequentemente, tem um desempenho ruim em novos dados que não foram vistos antes. Em outras palavras, o modelo se adapta excessivamente ao ruído e às variações aleatórias dos dados de treinamento em vez de aprender os padrões subjacentes que seriam relevantes para novos dados.</p>

<p align="justify">Em geral, o primeiro modelo usando a técnica de "over-sampling" com SMOTE apresentou indicadores gerais mais satisfatórios, porém é importante avaliar a capacidade desse modelo em lidar com dados novos e desconhecidos. Já o segundo modelo treinado pela técnica de validação cruzada (cross-validation) pode ser considerado uma boa opção para estimar o desempenho do modelo em dados novos e desconhecidos.</p>
